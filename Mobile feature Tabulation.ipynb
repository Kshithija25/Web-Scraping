```python
Model=[]
Display=[]
Processor=[]
Front_Camera=[]
Resolution=[]
RAM=[]
OSAndroid=[]
Storage=[]
Rear_Camera=[] 
Battery_Capacity=[]
```

```python
for i in range(0,len(url)):
    data=requests.get(url[i])
    data_soup=BeautifulSoup(data.text)
    model_tag=data_soup.find_all('div',id='productwdgttitle1')
    for i in model_tag:
        text=i.text
        res_join=''.join(text)
        Model.append(res_join)
    key_specs_tag=data_soup.find_all('div',class_='pd_colmn')
    key_specs_tag
    keyspecs=[]
    for i in key_specs_tag:
        text=i.text
        res_join=''.join(text)
        keyspecs.append(res_join)
    Display.append(keyspecs[0][7:])
    Processor.append(keyspecs[1][9:])
    Front_Camera.append(keyspecs[2][12:])
    Resolution.append(keyspecs[3][10:])
    RAM.append(keyspecs[4][3:])
    OSAndroid.append(keyspecs[5][10:])
    Storage.append(keyspecs[6][7:])
    Rear_Camera.append(keyspecs[7][11:])
    Battery_Capacity.append(keyspecs[8][16:])
```
